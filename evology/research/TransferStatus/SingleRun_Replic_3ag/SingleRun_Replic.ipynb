```python

import pandas as pd
import random 
import matplotlib
import matplotlib.pyplot as plt 
data = pd.DataFrame()
```

```python

run = pd.read_csv("/Users/aymericvie/Documents/GitHub/evology/evology/research/TransferStatus/SingleRun_Replic_3ag/data/simulation3ag_15k.csv")
ref = pd.read_csv("/Users/aymericvie/Documents/GitHub/evology/evology/data/replication_data_sample2.csv")
# ref = pd.read_csv("/Users/aymericvie/Documents/GitHub/evology/evology/data/replication_data_sample1.csv")

```

Now let us compare the simulations.
## Price

```python
''' Price '''
data['Price (Reference)'] = ref['Price']
data['Price (Simulation)'] = run['Price']
data['Gen'] = run['Gen']
data.plot(x="Gen", y = ["Price (Reference)", "Price (Simulation)"],
        kind="line", figsize=(15, 6), title='Comparison of price', 
        color = ['red','black'],
        style = ['-','-'])
plt.xlabel('Time (days)')
plt.ylabel('Price')
plt.savefig('figures/replic_price.png', dpi = 300)
plt.show() 
 
```

## Dividends


```python
''' Dividends '''
data = pd.DataFrame()
data['Dividends (Reference)'] = ref['Dividends']
data['Dividends (Simulation)'] = run['Dividends']
data['Gen'] = run['Gen']
data.plot(x="Gen", y = ["Dividends (Reference)", "Dividends (Simulation)"],
        kind="line", figsize=(15, 6), title='Comparison of dividends', 
        color = ['red','black'],
        style = [':','-'])
plt.xlabel('Time (days)')
plt.ylabel('Daily dividend')
plt.savefig('figures/replic_div.png', dpi = 300)
plt.show()

print(ref['Dividends'].std())
print(run['Dividends'].std())
```

## Volume



```python
''' Volume '''
data['Volume (Reference)'] = ref['Volume']
data['Volume (Simulation)'] = run['Volume']

fig, (ax1, ax2) = plt.subplots(2, 1, figsize = (10, 5), sharex = True, sharey = True)
fig.suptitle('Volume (Reference and Simulation)')
ax1.plot(data["Volume (Reference)"], label = 'Volume (Reference)')
ax2.plot(data["Volume (Simulation)"], label = 'Volume (Simulation)')
ax1.legend()
ax2.legend()
ax2.set_xlabel('Time (days)')
ax2.set_ylabel('Volume')
plt.savefig('figures/replic_vol.png', dpi = 300)
plt.show()
```

## Wealth

```python
''' Net asset value (i.e. asset long * price) '''
data['NT NAV (Reference)'] = ref['NT_nav']
data['NT NAV (Simulation)'] = run['NT_nav']
data['VI NAV (Reference)'] = ref['VI_nav']
data['VI NAV (Simulation)'] = run['VI_nav']
data['TF NAV (Reference)'] = ref['TF_nav']
data['TF NAV (Simulation)'] = run['TF_nav']
data['Gen'] = run['Gen']
data.plot(x="Gen", y = ["NT NAV (Reference)", "NT NAV (Simulation)","VI NAV (Reference)", 
    "VI NAV (Simulation)","TF NAV (Reference)", "TF NAV (Simulation)"],
        kind="line", figsize=(15, 6), title='Comparison of agent net asset values', 
        color = ['r', 'r', 'g', 'g', 'b', 'b'],
        style = [':', '-', ':', '-', ':', '-'])
plt.xlabel('Time (days)')
plt.ylabel('Net asset value')
plt.savefig('figures/replic_nav.png', dpi = 300)
plt.show()

```

## Wealth shares

```python
data['WShare_NT (Reference)'] = data['NT NAV (Reference)'] / (data['NT NAV (Reference)'] + data['VI NAV (Reference)'] + data['TF NAV (Reference)'])
data['WShare_VI (Reference)'] = data['VI NAV (Reference)'] / (data['NT NAV (Reference)'] + data['VI NAV (Reference)'] + data['TF NAV (Reference)'])
data['WShare_TF (Reference)'] = data['TF NAV (Reference)'] / (data['NT NAV (Reference)'] + data['VI NAV (Reference)'] + data['TF NAV (Reference)'])

data['WShare_NT (Simulation)'] = data['NT NAV (Simulation)'] / (data['NT NAV (Simulation)'] + data['VI NAV (Simulation)'] + data['TF NAV (Simulation)'])
data['WShare_VI (Simulation)'] = data['VI NAV (Simulation)'] / (data['NT NAV (Simulation)'] + data['VI NAV (Simulation)'] + data['TF NAV (Simulation)'])
data['WShare_TF (Simulation)'] = data['TF NAV (Simulation)'] / (data['NT NAV (Simulation)'] + data['VI NAV (Simulation)'] + data['TF NAV (Simulation)'])

fig, (ax1, ax2) = plt.subplots(2, 1, figsize = (10, 5), sharex = True, sharey = True)
fig.suptitle('Strategy wealth share (Reference and Simulation)')
ax1.plot(data['WShare_NT (Reference)'], label = 'WShare_NT (Reference)')
ax1.plot(data['WShare_VI (Reference)'], label = 'WShare_VI (Reference)')
ax1.plot(data["WShare_TF (Reference)"], label = 'WShare_TF (Reference)')
ax2.plot(data["WShare_NT (Simulation)"], label = 'WShare_NT (Simulation)')
ax2.plot(data["WShare_VI (Simulation)"], label = 'WShare_VI (Simulation)')
ax2.plot(data["WShare_TF (Simulation)"], label = 'WShare_TF (Simulation)')
ax1.legend()
ax2.legend()
ax2.set_xlabel('Time (days)')
ax2.set_ylabel('Wealth shares')
plt.savefig('figures/replic_wshare.png', dpi = 300)
plt.show()
```
